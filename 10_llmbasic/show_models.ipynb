```python
# 必要なモジュールをインポート
import os
from dotenv import load_dotenv
from openai import OpenAI

# .env を読み込む
load_dotenv()

# OpenAI クライアントの生成（APIキーは .env から取得）
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# モデルの一覧を取得
models = client.models.list()

# モデル一覧を表示
for model in models:
    print(model.id)
```
