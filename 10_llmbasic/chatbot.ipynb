```python
# 必要なモジュールをインポート
import os
from dotenv import load_dotenv
from openai import OpenAI
from pprint import pprint

# .env を読み込む
load_dotenv()

# OpenAI APIクライアントを生成（.env のキーを使う）
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# モデル名
MODEL_NAME = "gpt-4o-mini"

print("設定完了:", MODEL_NAME)
```

```python
# メッセージを格納するリスト
messages=[]

while(True):
    # ユーザーからの質問を受付
    message = input("メッセージを入力:")
    # 質問が入力されなければ終了
    if message.strip()=="":
        break
    print(f"質問:{message}")

    # メッセージにユーザーからの質問を追加
    messages.append({"role": "user", "content": message.strip()})
    # やりとりが8を超えたら古いメッセージから削除
    if len(messages) > 8:
        del_message = messages.pop(0)

    # APIへリクエスト
    stream = client.chat.completions.create(
        model=MODEL_NAME,
        messages=messages,
        stream=True,
    )

    # 言語モデルからの回答を表示
    response_message = ""
    for chunk in stream:
        if chunk.choices:
            next = chunk.choices[0].delta.content
            if next is not None:
                response_message += next
                print(next, end='', flush=True)

    # メッセージに言語モデルからの回答を追加
    messages.append({"role": "assistant", "content": response_message})

print("\n---ご利用ありがとうございました！---")
```

```python
# # 会話履歴（メッセージリスト）の初期化
messages = [
    {"role": "system", "content": "あなたは丁寧で親切に答えるアシスタントです。"}
]
```

```python
from IPython.display import display

while True:
    # ユーザーからの質問を受付
    message = input("メッセージを入力:")
    
    # 空なら終了
    if message.strip() == "":
        print("チャットを終了します。")
        break

    display(f"質問: {message}")
```

```python
# メッセージにユーザーからの質問を追加
messages.append({"role": "user", "content": message.strip()})

# やりとりが8件を超えたら古いメッセージから削除（system以外）
if len(messages) > 8:
    messages.pop(1)  # index 0 の system は残す
```

```python
stream = client.chat.completions.create(
    model=MODEL_NAME,
    messages=messages,
    stream=True,
)
```

```python
response_message = ""

for chunk in stream:
    # choices が空の場合はスキップ
    if not chunk.choices:
        continue

    delta = chunk.choices[0].delta

    # delta や delta.content が None の場合はスキップ
    if delta is None or delta.content is None:
        continue

    text = delta.content
    response_message += text
    print(text, end="", flush=True)
```

```python
# AI の返答が空でなければ履歴に追加
if response_message.strip():
    messages.append({"role": "assistant", "content": response_message})
```
